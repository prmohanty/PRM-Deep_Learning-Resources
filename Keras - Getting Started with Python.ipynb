Create 'Predictor Feature Set ' & 'Responce Feature Set'

```python
X = iris.values[:, :4]
y = iris.values[:, 4]
```

Split the Data into Training Set & Testing Set 

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=0)
```

Lets first create and test a ML model built with Scikit Learn

```python
LogReg = LogisticRegressionCV()
LogReg.fit(train_X, train_y)
```

```python
LogReg.score(test_X, test_y)
```

Now lets build a prediction model using Keras
=============================================

Keras is a high-level neural network library created by François Chollet at Google.

Step 1 
      scikit-learn's classifiers accept string labels, e.g. "setosa". But keras requires that labels be one-hot-encoded. 

```python
def one_hot_encode_object_array(arr):
    '''One hot encode a numpy array of objects (e.g. strings)'''
    uniques, ids = np.unique(arr, return_inverse=True)
    return np_utils.to_categorical(ids, len(uniques))

train_y_ohe = one_hot_encode_object_array(train_y)
test_y_ohe = one_hot_encode_object_array(test_y)
```

```python
train_y_ohe
```

```python
Keras_model = Sequential()
```

```python
# The next two lines define the size input layer (input_shape=(4,), 
# and the size and activation function of the hidden layer

Keras_model.add(Dense(16, input_shape=(4,)))
Keras_model.add(Activation('sigmoid'))
```

```python
# and these next lines defines the size and activation function of the ouput layer.

Keras_model.add(Dense(3))
Keras_model.add(Activation('softmax'))
```

```python
# And finally we specify the optimization strategy and the loss function to optimize. 
# We also instruct the model to calculate accuracy as it works.

Keras_model.compile(optimizer='adam', 
                    loss='categorical_crossentropy', 
                    metrics=["accuracy"])
```

```python
Keras_model.fit(train_X, train_y_ohe, epochs=100, batch_size=1, verbose=0);
```

evaluate() returns the loss function and any other metrics we asked for when we compiled the model. In our case, we asked for accuracy, which we can compare to the accuracy we got from the .score() method of our scikit-learn LogisticRegressionCV classifier.

```python
loss, accuracy = Keras_model.evaluate(test_X, test_y_ohe, verbose=0)
```
